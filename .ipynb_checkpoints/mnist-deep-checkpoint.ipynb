```python
import torch
import torchvision
from torchvision import datasets, transforms
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import matplotlib.pyplot as plt
import numpy as np
```

```python
import cv2
import numpy as np
import torch

def extract_advanced_features(image):
    # Convert the image to a numpy array and apply threshold
    np_image = image.numpy().squeeze()
    _, thresh = cv2.threshold(np_image, 0.5, 1.0, cv2.THRESH_BINARY)

    # Find contours in the thresholded image
    contours, _ = cv2.findContours((thresh*255).astype(np.uint8), 
                                   cv2.RETR_TREE, 
                                   cv2.CHAIN_APPROX_SIMPLE)

    # Count the number of closed contours which may correspond to loops
    loop_count = sum(cv2.contourArea(cnt) > 15 for cnt in contours)

    return torch.tensor([float(loop_count)])
```

```python
class AdvancedInfusionNet(nn.Module):
    def __init__(self):
        super(AdvancedInfusionNet, self).__init__()
        # Pathway for image data
        self.conv1 = nn.Conv2d(1, 32, kernel_size=5)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=5)
        # Pathway for advanced features
        self.adv_feature_fc = nn.Linear(1, 10)  # Assuming 1 advanced feature for now
        # Combined pathway
        self.fc1 = nn.Linear(1024 + 10, 128)  # Combine features from both pathways
        self.fc2 = nn.Linear(128, 10)

    def forward(self, image, adv_features):
        # Image pathway
        x1 = F.relu(F.max_pool2d(self.conv1(image), 2))
        x1 = F.relu(F.max_pool2d(self.conv2(x1), 2))
        x1 = x1.view(-1, 1024)
        # Advanced features pathway
        x2 = F.relu(self.adv_feature_fc(adv_features))
        # Combine pathways
        x = torch.cat((x1, x2), 1)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return F.log_softmax(x, dim=1)

net = AdvancedInfusionNet()
```

```python
transform = transforms.Compose([transforms.ToTensor(),
                                transforms.Normalize((0.5,), (0.5,))])

trainset = datasets.MNIST('', download=True, train=True, transform=transform)
testset = datasets.MNIST('', download=True, train=False, transform=transform)

trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True)
testloader = torch.utils.data.DataLoader(testset, batch_size=64, shuffle=False)
```

```python
criterion = nn.NLLLoss()
optimizer = optim.SGD(net.parameters(), lr=0.003, momentum=0.9)
```

```python
optimizer = torch.optim.SGD(net.parameters(), lr=0.003, momentum=0.9)

epochs = 10
for epoch in range(epochs):
    running_loss = 0
    for images, labels in trainloader:
        adv_features = torch.stack([extract_advanced_features(img) for img in images])

        optimizer.zero_grad()
        output = net(images, adv_features)
        loss = F.nll_loss(output, labels)

        loss.backward()
        optimizer.step()

        running_loss += loss.item()
    print(f"Epoch {epoch+1} - Training loss: {running_loss/len(trainloader)}")
```

```python
# get one batch from the test set
dataiter = iter(testloader)
images, labels = next(dataiter)

# get the advanced features
adv_features = torch.stack([extract_advanced_features(img) for img in images])

# get predictions
preds = torch.exp(net(images, adv_features))

# plot the images in the batch, along with predicted and true labels
fig = plt.figure(figsize=(25, 4))
with torch.no_grad():
    for idx in np.arange(20):
        ax = fig.add_subplot(2, 20//2, idx+1, xticks=[], yticks=[])
        ax.imshow(images[idx].squeeze(), cmap='gray')
        ax.set_title("{0}, {1:.1f}%\n(label: {2})".format(
            np.argmax(preds[idx]), 
            torch.max(preds[idx])*100, 
            labels[idx]),
                     color=("green" if np.argmax(preds[idx])==labels[idx].item() else "red"))
    
    
```

```python
def evaluate_model_accuracy(model, testloader):
    correct = 0
    total = 0
    with torch.no_grad():  # No need to track gradients during evaluation
        for images, labels in testloader:
            # For each batch in the test loader
            adv_features = torch.stack([extract_advanced_features(img) for img in images])
            outputs = model(images, adv_features)  # Get model predictions
            _, predicted = torch.max(outputs.data, 1)  # Find the max value
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    accuracy = 100 * correct / total
    return accuracy

# Calculate accuracy
model_accuracy = evaluate_model_accuracy(net, testloader)
print(f'Model Accuracy on Test Data: {model_accuracy}%')
```

```python
def extract_advanced_features_with_visualization(image):
    # Convert the image to a numpy array and apply threshold
    np_image = image.numpy().squeeze()
    _, thresh = cv2.threshold(np_image, 0.5, 1.0, cv2.THRESH_BINARY)

    # Find contours in the thresholded image
    contours, _ = cv2.findContours((thresh*255).astype(np.uint8), 
                                   cv2.RETR_TREE, 
                                   cv2.CHAIN_APPROX_SIMPLE)

    # Draw contours on the image for visualization
    contour_img = cv2.cvtColor((thresh*255).astype(np.uint8), cv2.COLOR_GRAY2BGR)
    cv2.drawContours(contour_img, contours, -1, (0, 255, 0), 1)

    # Count the number of closed contours
    loop_count = sum(cv2.contourArea(cnt) > 15 for cnt in contours)

    return torch.tensor([float(loop_count)]), contour_img
```

```python
import matplotlib.pyplot as plt

# Get some random images from the test set
dataiter = iter(testloader)
images, labels = next(dataiter)

# Visualize the feature extraction process
fig, axes = plt.subplots(len(images), 2, figsize=(10, len(images)*2))
if len(images) == 1:  # If there's only one example, axes is not a list
    axes = [axes]

for i, image in enumerate(images):
    # Extract advanced features and get the visualization
    _, contour_img = extract_advanced_features_with_visualization(image)
    
    # Original image
    axes[i][0].imshow(image.numpy().squeeze(), cmap='gray')
    axes[i][0].set_title(f'Original Image: {labels[i].item()}')
    axes[i][0].axis('off')

    # Contour image
    axes[i][1].imshow(contour_img, cmap='gray')
    axes[i][1].set_title(f'Processed for Feature Extraction')
    axes[i][1].axis('off')

plt.tight_layout()
plt.show()
```
